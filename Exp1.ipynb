# Exp-1: Implementation of Classification with Multilayer Perceptron using Scikit-learn with MNIST Dataset.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X, y = load_digits(return_X_y=True)
y = y.astype(int)
X /= 255.0 # Normalize Gray image to floats
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
m = MLPClassifier(hidden_layer_sizes=(64, 32, 16), max_iter=100)
m.fit(X_train, y_train)
```

```python
y_pred = m.predict(X_test)
len(y_pred)
```

```python
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
conf_mtx = confusion_matrix(y_test, y_pred)
print(conf_mtx)
```

```python
# optioinally plotting confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(conf_mtx, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
```
